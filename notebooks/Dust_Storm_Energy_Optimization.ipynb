# ?? RedPlanet: Dust Storm Energy Optimization
This notebook simulates a 30-day survival scenario for a Mars habitat encountering a major dust storm.

```python
import sys
sys.path.append('../src/eclss_energy_manager')
from energy_grid import EnergyGrid
import matplotlib.pyplot as plt

grid = EnergyGrid()
history = []

for day in range(1, 31):
    is_storm = 10 <= day <= 20
    res = grid.simulate_day(is_storm)
    res['day'] = day
    res['is_storm'] = is_storm
    history.append(res)

days = [h['day'] for h in history]
battery = [h['battery'] for h in history]
generation = [h['generation'] for h in history]
demand = [h['demand'] for h in history]
shed_count = [h['shed_count'] for h in history]

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(days, generation, label='Generation (kW)', color='orange')
plt.plot(days, demand, label='Demand (kW)', color='blue')
plt.fill_between(days, 0, 200, where=[h['is_storm'] for h in history], alpha=0.2, color='red', label='Dust Storm')
plt.title('Energy Generation vs Demand')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(days, battery, label='Battery Level (kWh)', color='green')
plt.ylabel('kWh')
plt.xlabel('Day')
plt.legend()
plt.tight_layout()
plt.show()
```
